```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Parametre
mean = 10
std_dev = 5
num_samples = 1500

# 2. Generer tilfældige x- og y-koordinater
x = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
y = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# 3. Opret en DataFrame
df_points = pd.DataFrame({'x': x, 'y': y})

# 4. Visualisering (Scatter Plot)
plt.figure(figsize=(10, 8))
sns.set_theme(style="darkgrid")

# Lav scatter plottet
sns.scatterplot(data=df_points, x='x', y='y', alpha=0.5, color='darkblue')

# Tilføj titler og labels
plt.title(f'Scatter Plot of {num_samples} Random Points (Normal Distribution)', fontsize=15)
plt.xlabel('X-akse (Mean=10, Std=5)', fontsize=12)
plt.ylabel('Y-akse (Mean=10, Std=5)', fontsize=12)

# Tilføj hjælpelinjer for gennemsnittet
plt.axvline(x=mean, color='red', linestyle='--', label='Mean (10)')
plt.axhline(y=mean, color='red', linestyle='--')

plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd

# 1. Del dit array (df_points) op i to halvdele
# Da vi har 1500 punkter, bliver hver halvdel 750 punkter
half = len(df_points) // 2
first_half = df_points.iloc[:half].values
second_half = df_points.iloc[half:].values

# 2. Beregn de euklidiske afstande
# Vi bruger Numpy's linalg.norm, som er den hurtigste og mest præcise måde
distances = np.linalg.norm(first_half - second_half, axis=1)

# 3. Gem resultatet i en ny DataFrame for overblik
df_distances = pd.DataFrame({
    'Distance': distances
})

# 4. Vis de 5 første resultater og statistikken
print("De første 5 euklidiske afstande:")
print(df_distances.head())

print("\nStatistik over afstandene:")
print(df_distances.describe())
```

```python
import numpy as np

# 1. Brug pandas' egen metode til at finde unikke værdier og sortere dem
# Vi gemmer dem i en variabel, der IKKE hedder 'sorted', 'min' eller 'max'
unique_array = df['DOLocationID'].dropna().unique()
unique_array.sort() # Sorterer arrayet direkte

# 2. Find min og max ved hjælp af metoderne på selve arrayet
min_id = int(unique_array.min())
max_id = int(unique_array.max())

# 3. Lav tjekket for manglende tal
expected_ids = set(range(min_id, max_id + 1))
actual_ids = set(unique_array)
missing_ids = expected_ids - actual_ids

# 4. Print resultatet
print(f"Interval: {min_id} til {max_id}")
print(f"Antal unikke ID'er: {len(actual_ids)}")

if not missing_ids:
    print("\n✅ Alle numre er i træk!")
else:
    print(f"\n❌ Der mangler {len(missing_ids)} numre.")
    print("De første 5 manglende:", sorted(list(missing_ids))[:5] if 'sorted' in globals() and callable(globals()['sorted']) else list(missing_ids)[:5])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Konverter DOLocationID til string for at få pænere kategorier på grafen
# Vi gemmer den som en ny midlertidig kolonne i dit grupperede datasæt
mean_dist_by_id['DOLocationID_str'] = mean_dist_by_id['DOLocationID'].astype(str)

# 2. Opsætning af visualisering
plt.figure(figsize=(18, 8))
sns.set_theme(style="whitegrid")

# 3. Lav Bar Plot - vi bruger barplot, da vi har beregnede værdier
# Vi viser kun hver 5. label på x-aksen for at undgå at teksten overlapper
ax = sns.barplot(x='DOLocationID_str', y='trip_distance', data=mean_dist_by_id,
                 palette='viridis', hue='DOLocationID_str', legend=False)

# 4. Tilføj titler og labels
plt.title('Gennemsnitlig Tur-distance fordelt på Drop-off ID (Kategorisk sorteret)', fontsize=16)
plt.xlabel('Drop-off Lokation ID (som tekst)', fontsize=12)
plt.ylabel('Gennemsnitlig Distance (miles)', fontsize=12)

# Vis kun hver 10. label for at få en ren og læselig x-akse
for i, t in enumerate(ax.get_xticklabels()):
    if i % 10 != 0:
        t.set_visible(False)

plt.xticks(rotation=90)
plt.show()

```
